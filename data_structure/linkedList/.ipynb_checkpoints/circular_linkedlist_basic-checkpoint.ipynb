## Basic circular linkedlist implementation
        

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class CircularLinkedList:
    def __init__(self):
        self.last = None
        
    def addToEmpty(self, data):
        temp = Node(data)
        self.last = temp
        self.last.next = self.last
        
        
    def addToStart(self, data):
        
        if self.last is None:
            self.addToEmpty(data)
        else:  
            temp = Node(data)
            temp.next = self.last.next
            self.last.next = temp
    
    def addToEnd(self,data):
        
        if self.last is None:
            self.addToEmpty(data)
        else:
            temp = Node(data)
            temp.next = self.last.next
            self.last.next = temp
            self.last = temp
            
    def addAfter(self, data, item):
        if self.last is None:
            return None
        p = self.last.next 
        
        while p:
            if p.data == item:
                
                temp = Node(data)
                temp.next = p.next
                p.next = temp
                
                if p == self.last:
                    self.last = temp
                return self.last
                
            p = p.next
        if p == self.last.next:
            raise ValueException("item not found")
        
    def traverse(self):
        if self.last is None:
            print('empty linked list')
            return
        
        temp = self.last.next 
        while temp:
            print(temp.data)
            temp = temp.next
            if temp == self.last.next:
                break
                
```

```python
llist = CircularLinkedList() 

last = llist.addToEmpty(6) 
last = llist.addToStart(4) 
last = llist.addToStart(2) 
last = llist.addToEnd(8) 
last = llist.addToEnd(12) 
last = llist.addAfter(10,8) 

llist.traverse()
```

```python
class CircularLinkedList_Head:
    def __init__(self):
        self.head = None
      
    def push(self, data): 
        ptr1 = Node(data) 
        temp = self.head 
          
        ptr1.next = self.head 
  
        # If linked list is not None then set the next of 
        # last node 
        if self.head is not None: 
            while(temp.next != self.head): 
                temp = temp.next 
            temp.next = ptr1 
  
        else: 
            ptr1.next = ptr1 # For the first node 
  
        self.head = ptr1 
    
    
    def sortedInsert(self, new_node):
        """
        3 cases:
        1. Empty linksedlist
        2. new data is smaller than originial head, insert as head
        3. insert in between
        """
        cur = self.head
        if cur is None:
            new_node.next = new_node
            self.head = new_node
            
        
        
        elif new_node.data < self.head.data:
            while cur.next != self.head:
                cur = cur.next
                
            new_node.next = cur.next
            cur.next = new_node
            self.head = new_node
        else:
            while cur.next !=self.head and cur.next.data < new_node.data:
                cur = cur.next
                
            new_node.next = cur.next
            cur.next = new_node
            
    def splitList(self, list1, list2):
        """
        steps
        1. Use toitorse and hare method to find the split point
        2. if even number of nodes, then fast.next.next = head
           if odd number of nodes, then fast.next = head
        3. appoint head for two lists
        4. Make two list circular
        """
        
        if self.head == None:
            return None
     
        fast = self.head
        slow = self.head
        while fast.next is not self.head and fast.next.next is not self.head:
            slow = slow.next
            fast = fast.next.next
            
        if fast.next.next == self.head:
            fast = fast.next
            
        list1.head = self.head
        
        if self.head.next != self.head:
            list2.head = slow.next
            
        fast.next = slow.next
        slow.next = self.head
    def printList(self): 
        temp = self.head 
        if self.head is not None: 
            while(True): 
                print("%d" %(temp.data))
                temp = temp.next
                if (temp == self.head): 
                    break 
```

```python
head = CircularLinkedList_Head()  
head1 = CircularLinkedList_Head() 
head2 = CircularLinkedList_Head() 
  
head.push(12) 
head.push(56) 
head.push(2) 
head.push(11) 
  
print("Original Circular Linked List")
head.printList() 
  
# Split the list  
head.splitList(head1 , head2) 
  
print("\nFirst Circular Linked List")
head1.printList() 
  
print("\nSecond Circular Linked List")
head2.printList()
```

```python
arr = [12, 56, 2, 11, 1, 90] 
  
list_size = len(arr) 
  
# start with empty linked list 
start = CircularLinkedList_Head() 
  
# Create linked list from the array arr[] 
# Created linked list will be 1->2->11->12->56->90 
for i in range(list_size): 
    temp = Node(arr[i]) 
    start.sortedInsert(temp) 
  
start.printList()
```
