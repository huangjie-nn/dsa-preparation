## Singly LinkedList manipulation

```python
class Node:
        def __init__(self,data):
            self.data = data
            self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def push(self, data):
        node = Node(data)
        temp = self.head
        if temp is None:
            self.head = node
        node.next = temp
        self.head = node
        
    def printList(self):
        node = self.head
        
        while node is not None:
            print(node.data)
            node = node.next
        
    def swapNodes(self, x, y):
        if x == y:
            return
        
        cur = self.head
        
        """
        finding prev node of x and y in a single loop
        """
        a = None
        b = None
        while cur.next is not None:
            if cur.next.data == x:
                a = cur
            
            if cur.next.data == y:
                b = cur
        
            cur = cur.next
            
        if a is not None and b is not None:
            temp = a.next
            a.next = b.next
            b.next = temp
            
            temp = a.next.next
            a.next.next = b.next.next
            b.next.next = temp
            
            return self.head
        
    def reverse(self):
        prev = None
        curr = self.head

        while curr is not None:
            nex = curr.next
            curr.next = prev
            prev = curr 
            curr = nex

        self.head = prev

    def reverseInGroup(self, head, k):
        prev = None
        curr = head
        cnt = 0
        while curr is not None and cnt < k:
            """
            steps are standard
            1. allocate the next
            2. reverse the pointer
            3. move prev and curr pointer
            """
            nex = curr.next
            curr.next = prev
            prev = curr
            curr = nex
            cnt += 1
        
        """
        final allocation of the head, being in group or as a whole
        """
        if nex is not None:
            head.next = self.reverseInGroup(nex, k)
        return prev
    
    def rotateByK(self,k):
        """
        3 steps
        1. find kth_node
        2. find last node
        3. point last node to the head
        4. point new head to the node next to kth_node
        5. kth_node next point to null
        
        """
        curr = self.head
        cnt = 1
        while cnt <k and curr is not None:
            curr = curr.next
            cnt += 1
            
        if curr is None:
            return 
        
        kth_node = curr
        
        while curr.next is not None:
            curr = curr.next
            
        curr.next = self.head
        self.head = kth_node.next
        kth_node.next = None
        
```

```python
## trying swapNode
l = LinkedList()
# The constructed linked list is:  
# 1.2.3.4.5.6.7  
l.push(7)  
l.push(6)
l.push(5)
l.push(4)
l.push(3)
l.push(2)
l.push(1)

print("Linked list before calling swapNodes() ")  
l.printList()  
  
l.swapNodes(6, 3)  
  
print("Linked list after calling swapNodes() ")  
l.printList()
```

```python
## trying out reverse
llist = LinkedList() 
llist.push(20) 
llist.push(4) 
llist.push(15) 
llist.push(85) 
  
print("Given Linked List")
llist.printList() 
llist.reverse() 
print("\nReversed Linked List")
llist.printList() 
```

```python
## trying out reverse in group
llist = LinkedList() 
llist.push(9) 
llist.push(8) 
llist.push(7) 
llist.push(6) 
llist.push(5) 
llist.push(4) 
llist.push(3) 
llist.push(2) 
llist.push(1) 
  
print("Given linked list")
llist.printList() 
llist.head = llist.reverseInGroup(llist.head, 3) 
  
print("\nReversed Linked list")
llist.printList()
```

```python
## trying rotation
llist = LinkedList() 
  
# Create a list 10->20->30->40->50->60 
for i in range(60, 0, -10): 
    llist.push(i) 
  
print("Given linked list")
llist.printList() 
llist.rotateByK(4) 
  
print("\nRotated Linked list")
llist.printList() 
```
