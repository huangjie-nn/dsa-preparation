## Binary indexed tree


We have an array [0...n-1]. We would like to 

1) compute the sum of the first i element
2) update the value of a specified element of the array arr[i] where 0<=i<=n-1

```python

def getSum(bitTree, i):
    s = 0
    i += 1
    while i>0:
        s += bitTree[i]

        i -= i & (-i)
    return s

def updateTree(bitTree, i, v):
    i += 1
    while i <= len(bitTree):
        bitTree[i] += v

        i += i & (-i)
        
def constructTree(arr):
    tree = [0] * (len(arr)+1)
    for idx in range(len(arr)):

        updateTree(tree, idx, arr[idx])

    return tree

freq = [2, 1, 1, 3, 2, 3, 4, 5, 6, 7, 8, 9] 
tree = constructTree(freq)
getSum(tree, 3)
```
