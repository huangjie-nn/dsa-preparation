## Find the first cicular tour that visits all petrol pumps



```python
class PetrolPump:
    def __init__(self, petrol, distance):
        self.petrol = petrol
        self.distance = distance
        
def solution(arr):
    n = len(arr)
    start = 0
    end = 1
    curr = arr[0].petrol - arr[0].distance
    while end != start :
        while end != start and curr < 0:
            curr -= arr[start].petrol - arr[start].distance
            start = (start + 1) % n
            
            if start == 0:
                return -1
            
        curr += arr[end].petrol - arr[end].distance
        
        end = (end + 1) % n
        
    return start

arr = [PetrolPump(6,4), PetrolPump(3,6), PetrolPump(7,3)] 
solution(arr) 
```

## Sliding window max

```python
from collections import deque

def solution(arr, k):
    Q = deque()
    for i in range(k):
        while Q and arr[i] >= arr[Q[-1]]:
            Q.pop()
            
        Q.append(arr[i])
        
    n = len(arr)
    for i in range(k, n):
        while Q:
            print(f'{Q[0]}  ')
        
        while Q and Q[0] <= i-k:
            Q.popleft()
        
        while Q and arr[i] >= arr[Q[-1]]:
            Q.pop()
        Q.append(arr[i])
        
    print(Q[0])

arr = [12, 1, 78, 90, 57, 89, 56] 
k = 3
printMax(arr, len(arr), k)
        
        
```

## Generate binary number from 1 to n

```python
from queue import Queue

def solution(n):
    q = Queue()
    
    q.put('1')
    while n >0:
        n -= 1
        s1 = q.get()
        print(s1)
        
        s2 = s1
        q.put(s1+"0")
        q.put(s2+'1')
        
solution(10)
```
